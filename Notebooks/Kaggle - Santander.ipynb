```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
import sklearn
```

```python
#Importing the dataset
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
```

```python
input=train_set.iloc[:,2:202]
output=train_set.iloc[:,1]
input_test=test_set.iloc[:,1:201]
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_input = sc.fit_transform(input)
df_input=pd.DataFrame(scaled_input)
df_input.head()
```

```python
scaled_input_test = sc.fit_transform(input_test)
df_input_test=pd.DataFrame(scaled_input)
df_input_test.head()
```

```python
#Train model
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
```

```python
def build_model(optimizer):
    model = Sequential()
    model.add(Dense(units=100, kernel_initializer = 'uniform', activation = 'relu', input_dim=200))
    model.add(Dense(units=100, kernel_initializer = 'uniform', activation = 'relu'))
    model.add(Dense(units=1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    model.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn = build_model)
parameters = {'batch_size': [25,50],
             'epochs': [10, 20],
             'optimizer':['adam', 'rmsprop']}
```

```python
grid_search = GridSearchCV(estimator = model,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10)
start_time = time.time()
grid_search = grid_search.fit(scaled_input,output)
elapsed_time = (start_time - time.time())/60
```

```python
#Best parameters
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
#Build model with best parameters
best_model = Sequential()
best_model.add(Dense(units=100, kernel_initializer = 'uniform', activation = 'relu', input_dim=200))
best_model.add(Dense(units=100, kernel_initializer = 'uniform', activation = 'relu'))
best_model.add(Dense(units=1, kernel_initializer = 'uniform', activation = 'sigmoid'))
best_model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
best_model.fit(scaled_input,output, batch_size =25, epochs = 10)
```

```python
predictions = best_model.predict_proba(scaled_input_test)
```
