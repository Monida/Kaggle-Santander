## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix
import seaborn as sns
```

## Import the datasets

```python
train_set = pd.read_csv('../Data/train.csv')
test_set = pd.read_csv('../Data/test.csv')
```

## Identify x and y for the train and test sets

```python
x_train=train_set.iloc[:,2:202]
y_train=train_set.iloc[:,1]

x_test=test_set.iloc[:,1:201]
```

## Create a correlation matrix to find correlation between any variable

```python
corr_matrix = x_train.corr()
```

```python
corr_matrix.head(5)
```

## Plot the correlation coefficient heatmap without  normalizing the variables

```python
plt.figure(figsize=(30,30))
sns.set(font_scale=2)
sns.heatmap(corr_matrix,xticklabels=10,yticklabels=10, vmin=-0.01, vmax = 0.01)
plt.savefig('..\Figures\Corr_coeff_heatmap_nonorm.png')
```

## Plot the correlation coefficient heatmap normalizing the variables

```python
#x_train normalized by column
x_train_norm_col = (x_train-x_train.mean())/x_train.std()
corr_matrix_norm_col = x_train_norm_col.corr()
```

```python
corr_matrix_norm_col.head()
```

```python
plt.figure(figsize=(30,30))
sns.set(font_scale=2)
sns.heatmap(corr_matrix_norm_col,xticklabels=10,yticklabels=10, vmin=-0.01, vmax = 0.01)
plt.savefig('..\Figures\Corr_coeff_heatmap_norm.png')
```
